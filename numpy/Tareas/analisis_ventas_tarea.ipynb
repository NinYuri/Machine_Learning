# Análisis de ventas con NumPy

Supongamos que tienes datos de ventas de una tienda en diferentes meses y quieres analizarlos para obtener información clave, como el total de ventas, el promedio de ventas por mes y los meses con mejor desempeño.

## Paso 1. Importar NumPy y definir los datos


```python
import numpy as np

# Meses del año
meses = np.array(["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio",
                  "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"])

# Datos de ventas por cada mes
ventas = np.array([2300, 2500, 2600, 2100, 2900, 3100, 2800, 2700, 3000, 3300, 3500, 3700])

print("\033[1;34m" + "MESES" + "\033[0m")
print(meses, "\n")
print("\033[1;34m" + "DATOS DE VENTA POR MES" + "\033[0m")
print(ventas)
```

## Paso 2. Análisis de ventas con Numpy

Aplicaremos funciones clave para extraer información útil

### 1. Calcular el total y promedio de ventas al año.

```python
# Obtener el total de las ventas y el promedio, imprimirlas
print("\033[1;35m" + "Total de Ventas" + "\033[0m\n" + f"${ventas.sum():.2f}\n")

print("\033[1;35m" + "Promedio de Ventas" + "\033[0m\n" + f"${np.mean(ventas):.2f}")
```

### 2. Identificar el mes con más y menos ventas

```python
# Mes con ventas máximas
max = np.argmax(ventas)
print("\033[1;35m" + "Mes con Mayores Ventas" + "\033[0m\n" + f"{meses[max]} ${ventas[max]}\n")

# Mes con ventas mínimas
min = np.argmin(ventas)
print("\033[1;35m" + "Mes con Menores Ventas" + "\033[0m\n" + f"{meses[min]} ${ventas[min]}")
```

### 3. Determinar los meses con ventas superiores al promedio

```python
# Meses con Ventas mayores al promedio
print("\033[1;35m" + "Ventas Superiores al Promedio" + "\033[0m\n")

prom = np.mean(ventas)
# iterar sobre las ventas con sus índices
for con, a in enumerate(ventas):
    if a > prom:
        print(f"{meses[con]}: ${a:.2f}")
```

### 4. Aplicar un incremento del 10% en las ventas

```python
print("\033[1;35m" + "Ventas Proyectadas para el Próximo Año" + "\033[0m")

ventas_proxa = ventas + (ventas * 0.1)
print(ventas_proxa)
```

### 5. Graficar las ventas actuales y el incremento para el próximo año

```python
# Importar la librería para graficar matplotlib
import matplotlib.pyplot as plt

# Agregar las etiquetas
fig, ax = plt.subplots(figsize = (9,6))

ax.plot(meses, ventas, marker = '*', linestyle = '--', color = 'magenta', label = '2024')
ax.plot(meses, ventas_proxa, marker = '*', linestyle = '--', color = 'cyan', label = '2025')

ax.set_title("Tendencia de Ventas Mensuales", color = 'blue', size = 12)
ax.set_xlabel('Mes', color = 'purple')
ax.tick_params(axis = 'x', rotation = 45, labelsize = 8)
ax.set_ylabel('Ventas ($)', color = 'purple')

ax.legend(loc = 'lower right')
ax.grid()
plt.show()
```

### 7. Detectar anomalías con un gráfico de caja (Boxplot)

```python
# Importar la librería seaborn
import seaborn as sns

sns.boxplot(data = ventas)
plt.title('Distribución de Ventas (Detección de Anomalías)', color = 'blue', size = 12)
plt.ylabel('Ventas ($)', color = 'purple')
plt.grid()
plt.show()
```
