```python
import pandas as pd
```

```python
# leer y mostrar dataframe
df = pd.read_csv('StudentsPerformance.csv')

print("\033[1;34m" + "DATAFRAME" + "\033[0m\n")
df
```

# Crear Index

```python
import numpy as np
import random
```

```python
# Crear valores no repetitivos para los index
new_index = np.arange(0, 1000)

print("\033[1;35m" + "Index 0 - 999" + "\033[0m")
print(new_index)
```

```python
# Barajar/desordenar los índices
random.shuffle(new_index)

print("\033[1;35m" + "Desordenar Índices" + "\033[0m")
print(new_index)
```

```python
# Crear una nueva columna con los nuevos index
print("\033[1;35m" + "Nueva Columna" + "\033[0m\n")
df['new index'] = new_index
df
```

# set_index()

```python
# Fijar columna new_index como index
print("\033[1;35m" + "Fijar Columna new index como índice" + "\033[0m\n")
df.set_index('new index', inplace = True)
df
```

# sort_index()

```python
# Ordenar dataframe por new_index
print("\033[1;35m" + "Ordenar Dataframe" + "\033[0m\n")
df.sort_index()
```

```python
# orden descendiente por index
print("\033[1;35m" + "Ordenar Dataframe Descendiente" + "\033[0m\n")
df.sort_index(ascending = False)
```
